### 4(d): Model Building

#### Let us divide into X & Y sets for model building

```python
y_train = bike_sharing_train.pop('cnt')
X_train = bike_sharing_train
```

#### Initialing the Linear Regression variable and doing the fitting

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

#### We will use RFE approach to indentify useful columns and then iterate according, we have total 30 cols to handle
- RFE is popular because it is easy to configure and use and because it is effective at selecting those features (columns) in a training dataset that are more or most relevant in predicting the target variable.

```python
# Let us start with approx 50% features
rfe = RFE(lm, 15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
# Let us check the results now
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# let us see what all columns are eliminated
X_train.columns[~rfe.support_]
```

```python
# let us check the columns by the rfe support
col = X_train.columns[rfe.support_]
col
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

### 4(e): Model Building using STATS

```python
#Creating Reusable Functions:
#First Function to generate model
def generate_model(cols):
    X_train_sm = sm.add_constant(X_train[cols])
    lm = sm.OLS(y_train, X_train_sm).fit()
    print(lm.summary())
    return lm
```

```python
#Third Function to get LR parameters
def generate_LR_params(lr):
    return lr.params
```

```python
#Third Function to generate VIF
def generate_VIF(cols):
    df = X_train[cols]
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'],2)
    print(vif.sort_values(by='VIF',ascending=False))
```

```python
generate_VIF(X_train_rfe.columns)
```

```python
lr1 = generate_model(X_train_rfe.columns)
```

## Part 5: Final Model Interpretation

##### Hypothesis Testing:

Hypothesis testing states that:
H0:B1=B2=...=Bn=0
H1: at least one Bi!=0

lr_final model coefficients: 
- const              0.038646
- yr                 0.236853
- temp               0.547216
- windspeed         -0.177210
- season_summer      0.091523
- season_winter      0.119440
- mnth_sept          0.094638
- weathersit_good    0.092103  

From the lr_final model summary, it is evident that all our coefficients are not equal to zero which means We can *REJECT the NULL HYPOTHESIS*

##### F Statistics
F-Statistics is used for testing the overall significance of the Model: Higher the F-Statistics, more significant the Model is.

- F-statistic: 294.6
- Prob (F-statistic): 3.08e-173
- The F-Statistics value are 294.6 (which is greater than 1) and the p-value of all the variables are well below 0.05 it states that the overall model is significant.

## Best Fit Equal Model:

### cnt = 0.038646 + yr*(0.236853) + temp*(0.547216) + windspeed*(-0.177210) + season_summer*(0.091523) + season_winter*(0.119440) + mnth_sept*(0.094638) + weathersit_good*(0.092103 )

### Interpretation of Coefficients:
This is similar to equation: Y = B0 + B1X1 + B2X2 ...Bn*Xn where:
- features: 
    - X1=yr
    - X2=temp
    - X3=windspeed
    - X4=season_summer
    - X5=season_winter
    - X6=mnth_sept
    - X7=weathersit_good

- If Positive sign: A coefficients value of (B1,B2,B4,B5,B6,B) indicated that a unit increase in Independent variable(X1,X2,X4,X5,X6), increases the bike hire numbers by (B1,B2,B4,B5,B6) units.
- If Negative sign: A coefficients value of (B3) indicated that a unit increase in Independent variable(X3), decreases the bike hire numbers by (B3) units.
- const: The Constant value of 0.038646 indicated that, in the absence of all other predictor variables(i.e. when x1,x2...xn =0), The bike rental can still increase by 0.038646 units.

## There is No Multicollinearity between the predictor variables

```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_final.columns
vif['VIF'] = [variance_inflation_factor(X_train_final.values, i) for i in range(X_train_final.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

- From the VIF calculation we could find that there is no multicollinearity existing between the predictor variables, as all the values are within permissible range of below 5
